```python
from octans import Portal

from matplotlib import pyplot as plt

import numpy as np
```

# Create a portal from name

```python
portal = Portal.from_name("Kepler-9")
portal
```

## Get kepler light curves

```python
kepler_xlcs = portal.kepler()
len(kepler_xlcs)
```

```python
fig, ax = plt.subplots()

for i, kepler_xlc in enumerate(kepler_xlcs):
    kepler_xlc.plot(ax=ax);

ax.get_legend().remove()
plt.show()
```

## Get tess light curves

```python
tess_xlcs = portal.tess()
len(tess_xlcs)
```

```python
fig, ax = plt.subplots()

for i, tess_xlc in enumerate(tess_xlcs):
    tess_xlc.plot(ax=ax);

ax.get_legend().remove()
plt.show()
```

## Get K2 lightcurves

```python
k2_xlcs = portal.k2()
len(k2_xlcs)
```

## Get ASAS Data

```python
portal3 = Portal.from_name("xy leo")
portal3
```

```python
asas_xlcs = portal3.asas()
```

```python
fig, ax = plt.subplots()

for i, asas_xlc in enumerate(asas_xlcs):
    asas_xlc[asas_xlc.flux.value > min(asas_xlc.flux.value)].plot(ax=ax, marker=".", linestyle="none", label=f"MAG_{i}");

plt.show()
```
